````python
# Creating correlation matrix with absolute values
corr_matrix = df.corr().abs()
````

````python
# Select upper triangle of correlation matrix
upper_matrix = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
````

````python
# Find features with correlation greater than 0.80
col_drop = [column for column in upper_matrix.columns if any(upper_matrix[column] > 0.90)]
````

````python
#columns to be dropped are
col_drop
````

````python
df.drop(col_drop,axis=1,inplace=True)
````

````python
df.shape
````

````python
from sklearn.model_selection import train_test_split
````

````python
x=df.drop('Rock or Mine',axis=1)
y=df['Rock or Mine']
````

````python
y
````

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
````

**MODEL BUILDING ✅**

````python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
````

````python
lr=LogisticRegression()
````

````python
lr.fit(x_train,y_train)
````

````python
y_pred=lr.predict(x_test)
````

````python
print("Accuracy=",accuracy_score(y_pred,y_test))
````

````python
y_t1=lr.predict(x_train)
````

````python
print("Training Accuracy=",accuracy_score(y_t1,y_train))
````
